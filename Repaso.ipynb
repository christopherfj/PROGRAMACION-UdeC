```python
M  = int( input('Ingrese numero entero: (M): ') )
while M<0:
  M  = int( input('Error\nIngrese numero entero: (M): ') )

N  = int( input('Ingrese numero entero: (N): ') )
while N<0:
  N  = int( input('Error\nIngrese numero entero: (N): ') )

if M>N:
  piso = N
  cielo = M
  signo = '+'
else:
  piso = M
  cielo = N
  signo = '-'

delta = 0
while piso<cielo:
  piso = piso+1
  delta = delta+1

print(M, '-', N, '=', signo, delta)
```

```python
divisor = 1 #(1)
suma = 0

n = int( input('Ingrese entero (n) >0: '))
while n<=0:
  n = int( input('Error\nIngrese entero (n) >0: '))
#n>0

while divisor<n: #(2)

  if (n%divisor)==0:
    suma = suma+divisor

  divisor = divisor+1 #(3)

if n==suma:
  print(n, "perfecto")
else:
  print(n, "no es perfecto")
  
```

```python

factorial = 1

n = int( input('Ingrese entero (n) >=0: '))
while n<0:
  n = int( input('Error\nIngrese entero (n) >=0: '))

n_aux = n #(1)
while n_aux>0: #(2)
  factorial = factorial*n_aux
  n_aux = n_aux-1 #(3)

digitos = 0
ceros = 0
f_aux = factorial

while f_aux!=0:
  digitos = digitos+1
  digito = f_aux%10
  f_aux = f_aux//10
  if digito==0:
    ceros = ceros+1
print('ceros:', ceros)
```

```python
N = int( input('Ingrese entero (N): '))

# a o b
#a or b
#a y b
#a and b

h_prev = int( input('Ingrese entero (h): '))
saltos_arriba = 0
saltos_abajo = 0

i = 0 #(1)

while i<(N-1): #(2)

  h_act = int( input('Ingrese entero (h): '))

  if h_prev>h_act:
    saltos_abajo = saltos_abajo+1

  elif h_prev<h_act:
    saltos_arriba = saltos_arriba+1

  h_prev = h_act

  i = i+1 #(3)

print('arriba:', saltos_arriba)
print('abajo:', saltos_abajo)
```
