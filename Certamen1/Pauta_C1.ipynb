```python
#P1: inversión de un número entero
#variables
m = int(input('Ingrese 100<=m<=10**12: '))
while m<100 or m>10**12:
  m = int(input('Error\nIngrese 100<=m<=10**12: '))
i = 0
s = 0
#inversión de número
m_aux = m
while m_aux!=0:
   d = m_aux%10
   s = s*10+d
   m_aux = m_aux//10
#salida
print(s)
```

```python
#P2: terna de valor más alto
#variables
T = int(input('Ingrese T>0: '))
while T<=0:
  T = int(input('Error\nIngrese T>0: '))
mayor = 0
i = 0
#terna valor más alto
while i<T:
  A = float(input('Ingrese A: '))
  B = float(input('Ingrese B: '))
  O = int(input('Ingrese O {1,2,3,4}: '))
  while O<1 or O>4:
    O = int(input('Ingrese O {1,2,3,4}: '))
  if O==1:
    r = A+B
  elif O==2:
    r = A-B
  elif O==3:
    r = A*B
  elif O==4:
    r = A/B
  if i==0:
    mayor = r
  else:
    if r>mayor:
      mayor = r
  i = i+1
print(mayor)
```

```python
#P3: secuencia de Fibonacci y número Áureo
#variables
n = int(input('Ingrese n>=2: '))
while n<2:
  n = int(input('Error\nIngrese n>=2: '))
i = 0
#fibonacci (n)
fn_1 = 1
fn_2 = 0
while i<(n-2):
  fn = fn_1+fn_2 
  fn_2 = fn_1
  fn_1 = fn
  i = i+1
#fibonacci (n+1)
fn_mas = fn_1+fn_2 
#aureo
phi = (1+5**0.5)/2
phi_f = fn_mas/fn
#salida
print(abs(phi-phi_f))
```
