```python
#P2-a
def validar(msje, inf, sup):
    numero = int(input('Ingrese valor '+msje))
    if inf!=sup:
        while numero<inf or numero>sup:
            numero = int(input('Error\nIngrese valor '+msje))
    else:
        while numero<inf:
            numero = int(input('Error\nIngrese valor '+msje))
    return numero

def TLD(correos):
    TLDs = []
    for correo in correos:
        tld = correo.split('.')[-1]
        TLDs.append(tld)
    return TLDs

def elementos_unicos(lista):
    unicos = []
    for elem in lista:
        if elem not in unicos:
            unicos.append(elem)
    return unicos

def max_TLD(correos):
    TLDs = TLD(correos)
    unicos = elementos_unicos(TLDs) #list(set(TLDs))
    max_ = -1
    unico = ''
    for unico in unicos:
        cuenta = TLDs.count(unico)
        if cuenta>max_:
            max_ = cuenta
            tld = unico
    return tld, max_
        
n = validar('n>0: ', 1, 1)
i = 0
correos = []
while i<n:
    correo = input('Ingrese correo: ')
    correos.append(correo)
    i = i+1

#correos = ['hola@gmail.com', 'sec_docencia@inf.udec.cl', 'vicsanchez2021@udec.cl', \
#            'desconocido@hotmail.com', 'gboric@presidencia.gob.cl']
print( max_TLD(correos) )
```

```python
#P2-b
def verifica_minusculas(cadena):
    minusculas = 'abcdefghijklmn√±opqrstuvwxyz '
    verifica = True
    for c in cadena:
        if c not in minusculas:
            verifica = False
            break
    return verifica

def validar(msje, inf, sup):
    cadena = input('Ingrese palabra '+msje)
    if inf!=sup:
        while len(cadena)<inf or len(cadena)>sup or not verifica_minusculas(cadena):
            cadena = input('Error\nIngrese palabra '+msje)
    else:
        while len(cadena)<inf or not verifica_minusculas(cadena):
            cadena = input('Error\nIngrese palabra '+msje)
    return cadena
        
def jerigonza(cadena):
    vocales = 'aeiou'
    cadena_aux = ''
    for i in range(len(cadena)-1):
        c = cadena[i]
        if c in vocales and cadena[i+1] not in vocales:
            cadena_aux = cadena_aux+c+'p'+c
        else:
            cadena_aux = cadena_aux+c
    c = cadena[i+1]
    if c in vocales:
        cadena_aux = cadena_aux+c+'p'+c
    else:
        cadena_aux = cadena_aux+c
    return cadena_aux

#hola como estas
#estoy mal
#yo estoy bien
#que bueno
cadena = validar('1000>=len>=10: ', 10, 1000)
print( jerigonza(cadena) )
```
