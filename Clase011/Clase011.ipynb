```python
#P2
import numpy as np

def validar(msje, inf, sup):
    numero = int(input('Ingrese valor '+msje))
    if inf!=sup:
        while numero<inf or numero>sup:
            numero = int(input('Error\nIngrese valor '+msje))
    else:
        while numero<inf:
            numero = int(input('Error\nIngrese valor '+msje))
    return numero

def matriz_simetrica(m):
    x,y = m.shape
    s = 0
    for i in range(x):
        for j in range(y):
            if m[i][j] == m[j][i]:
                s = s+1
    if s==(x*y):
        print('La matriz es simétrica')
    else:
        print('La matriz no es simétrica')

#A = np.array( [ [4,9,6], [9,0,-7], [6,-7,-1] ])

N = validar('100>N>=2', 2, 99)
A = np.empty((N,N))
i = 0
while i<N:
    j = 0
    while j<N:
        valor = float( input(f'Ingrese valor {i,j}: ') )
        A[i,j] = valor
        j = j+1
    i = i+1

print(A)

matriz_simetrica(A)
x,y = A.shape
print( (A == A.T).sum()==(x*y) )
```

```python
#P3
import numpy as np

def validar(msje, inf, sup):
    numero = int(input('Ingrese valor '+msje))
    if inf!=sup:
        while numero<inf or numero>sup:
            numero = int(input('Error\nIngrese valor '+msje))
    else:
        while numero<inf:
            numero = int(input('Error\nIngrese valor '+msje))
    return numero

def triangular_superior(m):
    x,y = m.shape
    flag = True
    for i in range(1, x):
        if flag:
            for j in range(0,i):
                if m[i,j] != 0:
                    flag = False
                    break
        else:
            break
    if flag:
        print('La matriz es triangular superior')
    else:
        print('La matriz no es triangular superior')
            

#A = np.array( [ [4,9,6], [0,0,-7], [0,0,-1] ])

N = validar('100>N>=2', 2, 99)
A = np.empty((N,N))

for i in range(N):
    for j in range(N):
        valor = float( input(f'Ingrese valor {i,j}: ') )
        A[i,j] = valor

print(A)

triangular_superior(A)
```

```python
#P4
import numpy as np

def validar(msje, inf, sup):
    numero = int(input('Ingrese valor '+msje))
    if inf!=sup:
        while numero<inf or numero>sup:
            numero = int(input('Error\nIngrese valor '+msje))
    else:
        while numero<inf:
            numero = int(input('Error\nIngrese valor '+msje))
    return numero

def matriz_traspuesta(m):
    x,y = m.shape
    m_t = np.empty((y,x))
    for i in range(x):
        for j in range(y):
            m_t[j,i] = m[i,j]
    return m_t
            
    
#A = np.array([ [4,9], [-7,6], [0,-1] ])

n = validar('n>1: ', 2, 2)
m = validar('m>1: ', 2, 2)
A = np.empty((n,m))

for i in range(n):
    for j in range(m):
        valor = float( input(f'Ingrese valor {i,j}: ') )
        A[i,j] = valor

print( matriz_traspuesta(A) )
print(A.T)
```

```python
#P5
import numpy as np

def validar(msje, inf, sup):
    numero = int(input('Ingrese valor '+msje))
    if inf!=sup:
        while numero<inf or numero>sup:
            numero = int(input('Error\nIngrese valor '+msje))
    else:
        while numero<inf:
            numero = int(input('Error\nIngrese valor '+msje))
    return numero

def multiplica_matrices(m1,m2):
    x,y = m1.shape
    m1m2 = np.zeros((x,y))
    for i in range(x):
        for j in range(x):
            s = 0
            for k in range(y):
                s = s+m1[i,k]*m2[k,j]
            m1m2[i,j] = s 
    return m1m2

def idempotente(m):
    x,y = m.shape
    m2 = multiplica_matrices(m,m)
    s = 0
    for i in range(x):
        for j in range(y):
            if m[i,j]==m2[i,j]:
                s = s+1
    if s==(x*y):
        print('La matriz es idempotente')
    else:
        print('La matriz no es idempotente')
            
#A = np.array( [ [2, -1], [2, -1] ] )
 
n = validar('n>1: ', 2, 2)
A = np.empty((n,m))

for i in range(n):
    for j in range(n):
        valor = float( input(f'Ingrese valor {i,j}: ') )
        A[i][j] = valor
    
idempotente(A)
print( np.dot(A,A).sum()==A.sum() )
```
