```python
#P1
import numpy as np

def validar(msje, inf, sup):
    numero = int(input('Ingrese valor '+msje))
    if inf!=sup:
        while numero<inf or numero>sup:
            numero = int(input('Error\nIngrese valor '+msje))
    else:
        while numero<inf:
            numero = int(input('Error\nIngrese valor '+msje))
    return numero

def producto_escalar(u,v):
    s = 0
    N = len(u)
    for i in range(N):
        s = s+u[i]*v[i]
    return s

#A = np.array([4.0, 9.0, 6.0, -1.0, 3.0, -2.0])
#A = np.array( input('valor: ').split(' '), dtype=float )
#B = np.array([-3.0, 4.0, -7.0, 1.0, 8.0, -3.0])

n = validar('n>2: ', 3,3)
i = 0
A = []
B = np.empty(n) #alternativa a zeros, ones
while i<n:
    a = float(input('A: '))
    b = float(input('B: '))
    A.append(a)
    B[i] = b
    i = i+1
A = np.array(A)
print(A,B)

#Solución usando métodos (funciones) no numpy
print( producto_escalar(A,B) )
#Alternativa usando numpy: 
#Multiplicación punto a punto al ser array (array*array). Luego al hacer sum se suman todos los elementos del arreglo resultante
print( (A*B).sum() ) 
#Alternativa usando numpy:
#Multiplicación punto a punto usando solamente dot
print( np.dot(A,B) )
```

```python
#P2
import numpy as np

def validar(msje, inf, sup):
    numero = int(input('Ingrese valor '+msje))
    if inf!=sup:
        while numero<inf or numero>sup:
            numero = int(input('Error\nIngrese valor '+msje))
    else:
        while numero<inf:
            numero = int(input('Error\nIngrese valor '+msje))
    return numero

def matriz_simetrica(m):
    x,y = m.shape
    s = 0
    for i in range(x):
        for j in range(y):
            if m[i][j] == m[j][i]:
                s = s+1
    if s==(x*y):
        print('La matriz es simétrica')
    else:
        print('La matriz no es simétrica')

#A = np.array( [ [4,9,6], [9,0,-7], [6,-7,-1] ])

N = validar('100>N>=2: ', 2, 99)
A = np.empty((N,N))
i = 0
while i<N:
    j = 0
    while j<N:
        valor = float( input(f'Ingrese valor {i,j}: ') )
        A[i,j] = valor #equivalente a A[i][j] en notación de listas
        j = j+1
    i = i+1

print(A)

#Solución usando métodos (funciones) no numpy
matriz_simetrica(A)
#Alternativa usando numpy: 
#shape entrega las dimensiones del arreglo (en este caso, 2D)
x,y = A.shape
#.T entrega la traspuesta del arreglo
#Al hacer == se compara elemento a elemento y se asigna un True (1) o un False (0)
#Se compara la suma de este arreglo resultante con la cantidad de elementos
print( (A == A.T).sum()==(x*y) )
```

```python
#P3
import numpy as np

def validar(msje, inf, sup):
    numero = int(input('Ingrese valor '+msje))
    if inf!=sup:
        while numero<inf or numero>sup:
            numero = int(input('Error\nIngrese valor '+msje))
    else:
        while numero<inf:
            numero = int(input('Error\nIngrese valor '+msje))
    return numero

def triangular_superior(m):
    x,y = m.shape
    flag = True
    for i in range(1, x):
        if flag:
            for j in range(0,i):
                if m[i,j] != 0:
                    flag = False
                    break
        else:
            break
    if flag:
        print('La matriz es triangular superior')
    else:
        print('La matriz no es triangular superior')
            

#A = np.array( [ [4,9,6], [0,0,-7], [0,0,-1] ])

N = validar('100>N>=2: ', 2, 99)
A = np.empty((N,N))

#Otra forma de inicializar una matriz, en este caso, por filas
for i in range(N):
    #Es posible especificar el tipo del arreglo con dtype
    valores = np.array( input(f'Ingrese valores fila separados por un espacio: ').split(' '), dtype=float )
    A[i,:] = valores #fila i-ésima, todas las columnas (:)

print(A)

#Solución usando métodos (funciones) no numpy
triangular_superior(A)
```

```python
#P4
import numpy as np

def validar(msje, inf, sup):
    numero = int(input('Ingrese valor '+msje))
    if inf!=sup:
        while numero<inf or numero>sup:
            numero = int(input('Error\nIngrese valor '+msje))
    else:
        while numero<inf:
            numero = int(input('Error\nIngrese valor '+msje))
    return numero

def matriz_traspuesta(m):
    x,y = m.shape
    m_t = np.empty((y,x))
    for i in range(x):
        for j in range(y):
            m_t[j,i] = m[i,j]
    return m_t
            
    
#A = np.array([ [4,9], [-7,6], [0,-1] ])

n = validar('n>1: ', 2, 2)
m = validar('m>1: ', 2, 2)
A = np.empty((n,m))

for i in range(n):
    for j in range(m):
        valor = float( input(f'Ingrese valor {i,j}: ') )
        A[i,j] = valor

#Solución usando métodos (funciones) no numpy
print( matriz_traspuesta(A) )
#Alternativa usando numpy: 
#.T entrega la traspuesta del arreglo
print(A.T)
```
