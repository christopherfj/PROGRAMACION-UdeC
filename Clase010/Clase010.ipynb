```python
#P3
def verifica_caracter(cadena):
    verifica = True
    for c in cadena:
        if c not in '()':
            verifica = False
            break
    return verifica

def validar_cadena():
    cadena = input('Ingrese cadena: ')
    while not verifica_caracter(cadena):
        cadena = input('Error\nIngrese cadena: ')
    return cadena
    
def parentesis():
    cadena = validar_cadena()
    cadena_aux = cadena[:]
    while '()' in cadena_aux:
        cadena_aux = cadena_aux.replace('()', '')
    if cadena_aux == '':
        print('Expresión de paréntesis correctamente construida')
    else:
        print('Expresión de paréntesis incorrectamente construida')

#()(())()
#()(())()
#(()))(()
#())(()
#(()())
        
parentesis()
```

```python
#P4
def verifica_letras(cadena):
    verifica = True
    for c in cadena:
        if c not in ['A', 'C', 'G', 'T']:
            verifica = False
            break
    return verifica

def validar(msje, inf, sup):
    P = input(msje)
    while len(P)<inf or len(P)>sup or not verifica_letras(P):
        P = input('Error\n'+msje)
    return P

def infeccion():
    bacteria = validar('Bacteria, 100=>l>=1: ', 1, 100)
    l_original = len(bacteria)
    virus = validar('Virus, 100=>l>=1: ', 1, 100)
    while virus in bacteria:
        bacteria = bacteria.replace(virus, '')
    l_mod = len(bacteria)
    preservacion = 100*l_mod/l_original
    
    print( 'Bacteria:', bacteria, '(', round(preservacion,3), '% )' )

#GCTTTCGACGAT TCGA
#GATCGAGCTTCGAA TCGA
#GGTCTAGCTAAT, TCGA
infeccion()
```

```python
#P5
def unicas(palabras):
    palabras_u = []
    for palabra in palabras:
        if palabra not in palabras_u:
            palabras_u.append( palabra )
    return palabras_u

def twitter():
    palabras_tweets = []
    tweet = 'TWITTER'
    while tweet != '':
        tweet = input('Ingrese tweet: ')
        tweet = tweet.lower()
        palabras = tweet.split(' ')
        for palabra in palabras:
            if len( palabra )>=3:
                palabras_tweets.append( palabra )
    palabras_u = unicas(palabras_tweets) #list(set(palabras_tweets))
    max_ = 0
    tendencia = ''
    for palabra in palabras_u:
        frec = palabras_tweets.count( palabra )
        if frec>max_:
            max_=frec
            tendencia = palabra
    return tendencia
        
#Piñera presumiendo sus contratos con Pfizer fin al COVID
#Con un siete por ciento Piñera lidera el grupo de presidentes con menor respaldo en la historia republicana de Chile
#Nuevos brotes de covid en Los Ángeles
#Tercera ola de Covid en Europa
#Vacunas contra el Covid aplicadas en inglaterra usa y canadá
#Nueva cuarentena en región del Bío Bío por aumento de casos de covid positivo
#

print( twitter() )
```

```python
#P6
import itertools

def validar(msje, inf, sup):
    cadena = input('Ingrese palabra '+msje)
    if inf!=sup:
        while len(cadena)<inf or len(cadena)>sup:
            cadena = input('Error\nIngrese palabra '+msje)
    else:
        while len(cadena)<inf:
            cadena = input('Error\nIngrese palabra '+msje)
    return cadena

def anagrama(cadena): 
    N = len(cadena)
    permutaciones = list(itertools.permutations(cadena, N))
    M = int(len(permutaciones)/2)
    for i in range(M):
        print(str(i+1)+'.',''.join(permutaciones[i]),'\t', str(M+i+1)+'.', ''.join(permutaciones[M+i]))
    return permutaciones

cadena = validar('10>largo>1: ', 1, 10)
anagramas = anagrama(cadena)
posiciones = input('Ingrese posiciones: ').split(' ')
for pos in posiciones:
    print(''.join(anagramas[int(pos)-1]))
```
